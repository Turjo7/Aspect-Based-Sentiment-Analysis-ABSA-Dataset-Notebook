**Importing Libraries**

```python
import gradio as gr
import pandas as pd
from pyabsa import ATEPCCheckpointManager
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch.nn.functional as F
```

**Base Code for Aspect Extraction**

```python
# Initialize the aspect extractor
aspect_extractor = ATEPCCheckpointManager.get_aspect_extractor(
    checkpoint='english',
    auto_device=True
)

def extract_aspects(text):
    # Perform aspect extraction
    result = aspect_extractor.extract_aspect(
        inference_source=[text],
        pred_sentiment=True
    )
    # Extract and return the aspects
    return result[0]['aspect']
```

**Loading ABSA Models**

```python
absa_tokenizer = AutoTokenizer.from_pretrained("yangheng/deberta-v3-base-absa-v1.1")
absa_model = AutoModelForSequenceClassification.from_pretrained("yangheng/deberta-v3-base-absa-v1.1")
```

**Analyze Function of ABSA Sarcasm/Non-Sarcasm**

```python
def analyze_sentiment(text):
    # Extract aspects from the single string
    aspects = extract_aspects(text)

    if not aspects:
        return "No aspects found."

    aspect = " ".join(aspects)

    # Tokenize and get model predictions for the aspect-based sentiment analysis
    inputs = absa_tokenizer(f"[CLS] {text} [SEP] {aspect} [SEP]", return_tensors="pt")
    outputs = absa_model(**inputs)
    probs = F.softmax(outputs.logits, dim=1)
    probs = probs.detach().numpy()[0]

    # Define labels
    labels = ["Sarcastic", "Non Sarcastic"]

    # Determine the label with the highest probability
    max_prob_index = probs.argmax()
    max_prob_label = labels[max_prob_index]

    # Format the output
    result = f"The aspect '{aspect}' is {max_prob_label} with a probability of {probs[max_prob_index]:.4f}"
    return result
```

**Gradio Code**

```python
# Create the Gradio interface
iface = gr.Interface(
    fn=analyze_sentiment,
    inputs="text",
    outputs="text",
    title="Aspect-Based Sentiment Analysis (ABSA)",
    description="Enter a sentence to determine if it is sarcastic or non-sarcastic based on extracted aspects."
)

# Launch the Gradio interface
iface.launch()
```
